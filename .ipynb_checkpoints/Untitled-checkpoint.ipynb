```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load the data
data = pd.read_csv('cleaned_quarterly_data2.csv')
```

```python
# Extract the last two characters from the 'Quarter' column
data['Quarter'] = data['Quarter'].apply(lambda x: x[-2:])

# Verify the results
print(data[['Quarter']].head())
print(data[['Quarter']].tail())
```

```python
# Extract the 'Quarter' column for encoding
quarter_data = data[['Quarter']]

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop='first')  # dropping the first category to avoid dummy variable trap

# Fit and transform the 'Quarter' column
encoded_quarters = encoder.fit_transform(quarter_data)

# Convert the encoded array to a DataFrame
encoded_quarters_df = pd.DataFrame(encoded_quarters, columns=encoder.get_feature_names_out(['Quarter']))

# Combine the encoded columns back into the original dataframe
data = data.reset_index(drop=True)
data_encoded = pd.concat([data, encoded_quarters_df], axis=1)

# Verify the results
print(data_encoded.head())
print(data_encoded.tail())
```

```python
data_encoded.to_csv('cleaned_quarterly_data_encoded.csv', index=False)
```

```python
from category_encoders import TargetEncoder

# Create a copy of the data to apply target encoding
data_encoded_target = data_encoded.copy()

# Initialize the TargetEncoder
encoder = TargetEncoder(cols=['RegionName', 'StateName'])

# Fit and transform the 'RegionName' and 'StateName' columns based on 'Home_Index'
data_encoded_target[['RegionName', 'StateName']] = encoder.fit_transform(data_encoded[['RegionName', 'StateName']], data_encoded['Home_Index'])

# Verify the results
print(data_encoded_target[['RegionName', 'StateName']].head())
print(data_encoded_target[['RegionName', 'StateName']].tail())
```

```python
# Save the cleaned and target encoded data to a new CSV
data_encoded_target.to_csv('cleaned_quarterly_data_target_encoded.csv', index=False)
```

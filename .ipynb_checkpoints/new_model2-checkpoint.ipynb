```python
columns = ['Home_Index', 'RegionName_Target_Encoded' ,'StateName_Target_Encoded', 'SizeRank', 'Year_Quarter', '30_Yr_Mort_Rate', 'GDP', 'CPI', 'Unemp_Rate', 'Per_Inc', 'Population']
corr_matrix = df[columns].corr()
```

```python
corr_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot the correlation heatmap before handling zeros
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Reorder columns for each set
train_data = train_data[columns]
val_data = val_data[columns]
test_data = test_data[columns]


# Function to plot correlation heatmap
def plot_correlation_matrix(data, title):
    plt.figure(figsize=(12, 8))
    corr_matrix = data.corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
    plt.title(title)
    plt.show()

# Plot correlation heatmaps for each subset
plot_correlation_matrix(train_data, 'Correlation Matrix - Training Set')
plot_correlation_matrix(val_data, 'Correlation Matrix - Validation Set')
plot_correlation_matrix(test_data, 'Correlation Matrix - Test Set')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.DataFrame(df)
```

```python
df
```

```python
columns_to_standardize = ['SizeRank', 'GDP', 'CPI', 'Unemp_Rate', 'Per_Inc', 'Population']
```

```python
scaler = StandardScaler()
```

```python
# Assuming df is your training set for demonstration
train_data = df.copy()  
val_data = df.copy()  
test_data = df.copy()  

# Standardize the features in the training set
train_data[columns_to_standardize] = scaler.fit_transform(train_data[columns_to_standardize])

# Standardize the features in the validation and test sets
val_data[columns_to_standardize] = scaler.transform(val_data[columns_to_standardize])
test_data[columns_to_standardize] = scaler.transform(test_data[columns_to_standardize])
```

```python
print(train_data[columns_to_standardize].mean(axis=0))  # Should be close to 0
print(train_data[columns_to_standardize].std(axis=0))   # Should be close to 1
```

```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_quarterly_data_encoded.csv')
```

```python
df.drop(columns=['Quarter_Q4', 'Quarter_Q2', 'Quarter_Q3'])
```

```python
# Convert Quarter to numerical value
quarter_to_num = {'Q1': 0.0, 'Q2': 0.25, 'Q3': 0.5, 'Q4': 0.75}
df['Quarter_num'] = df['Quarter'].map(quarter_to_num)

# Combine Year and Quarter_num
df['Year_Quarter'] = df['Year'] + df['Quarter_num']

print(df)
```

```python
df
```

```python
# Drop the unwanted columns
df = df.drop(columns=['Quarter_Q2', 'Quarter_Q3', 'Quarter_Q4', 'Year', 'Quarter', 'Quarter_num'])

# Verify the columns are dropped
print(df)
```

```python
df.to_csv('year_quarter_combined.csv', index=False)
```

```python
# Calculate mean Home_Index for each RegionName
region_mean = df.groupby('RegionName')['Home_Index'].mean()

# Map the mean values to the DataFrame
df['RegionName_Target_Encoded'] = df['RegionName'].map(region_mean)

# Calculate mean Home_Index for each StateName
state_mean = df.groupby('StateName')['Home_Index'].mean()

# Map the mean values to the DataFrame
df['StateName_Target_Encoded'] = df['StateName'].map(state_mean)

# Drop original RegionName and StateName columns if no longer needed
df = df.drop(columns=['RegionName', 'StateName'])

print(df)
```

```python
df.to_csv('year_quarter_combined_w_target_encoded.csv', index=False)
```

```python
train_split_point = 2012.75
validation_split_point = 2018.5
```

```python
train_data = df[df['Year_Quarter'] <= train_split_point]
validation_data = df[(df['Year_Quarter'] > train_split_point) & (df['Year_Quarter'] <= validation_split_point)]
test_data = df[df['Year_Quarter'] > validation_split_point]
```

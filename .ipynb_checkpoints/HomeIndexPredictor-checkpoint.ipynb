```python
import pandas as pd
```

```python
# Load the data
file_path = 'Metro_zhvi_uc_sfrcondo_tier_0.33_0.67_month.csv'
monthly_data = pd.read_csv(file_path)

# Display basic information about the dataset
print(monthly_data.info())
```

```python
print(monthly_data.head())
```

```python
# Convert wide data to long format
monthly_data_long = pd.melt(monthly_data, 
                            id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName'],
                            var_name='Date', 
                            value_name='Home_Index')

# Convert 'Date' to datetime format
monthly_data_long['Date'] = pd.to_datetime(monthly_data_long['Date'])

# Extract year and quarter from the date
monthly_data_long['Year'] = monthly_data_long['Date'].dt.year
monthly_data_long['Quarter'] = monthly_data_long['Date'].dt.to_period('Q')

# Verify the long format data
print(monthly_data_long.head())
```

```python
# Replace zeros in Home_Index with NaN for accurate mean calculation
monthly_data_long['Home_Index'] = monthly_data_long['Home_Index'].replace(0, pd.NA)

# Group by relevant columns and calculate the mean, skipping NaN values
quarterly_data = monthly_data_long.groupby(['RegionID', 'RegionName', 'StateName', 'Year', 'Quarter']).agg(
    {'Home_Index': lambda x: x[x.notna()].mean(),
     'SizeRank': 'first',  # Assuming SizeRank remains the same within a region
     'RegionType': 'first'}).reset_index()

# Fill NaN values back with zeros if required for other columns (optional, depending on use case)
quarterly_data = quarterly_data.fillna(0)

# Verify the results
print(quarterly_data.head())
```

```python
quarterly_data.to_csv('newformat.csv', index=False)
```

```python
quarterly_data = quarterly_data[~((quarterly_data['Year'] == 1996) & (quarterly_data['Quarter'] == '1996Q1'))]
```

```python
quarterly_data = quarterly_data[~((quarterly_data['Year'] == 2024) & (quarterly_data['Quarter'] == '2024Q2'))]
```

```python
quarterly_data.to_csv('newformat.csv', index=False)
```

```python
import numpy as np

# Create a copy of the original DataFrame to apply rolling mean
qtr_data_rolling = qtr_data.copy()

# Verify data types and ensure numerical types for numerical columns
# Convert Home_Index to numeric to ensure correct data type for rolling mean
qtr_data_rolling['Home_Index'] = pd.to_numeric(qtr_data_rolling['Home_Index'], errors='coerce')

# Apply rolling mean to Home_Index to handle zeros
# Using a window size of 4 for quarterly data (1 year)
# Replace zeros with NaN before applying rolling mean
qtr_data_rolling['Home_Index_Rolling'] = qtr_data_rolling.groupby('RegionID')['Home_Index'].transform(lambda x: x.replace(0, np.nan).rolling(window=4, min_periods=1).mean())

# Replace any remaining NaNs (if the first few points are NaN) with the mean of the series
# This ensures no NaN values remain after the rolling mean operation
qtr_data_rolling['Home_Index_Rolling'].fillna(qtr_data_rolling['Home_Index_Rolling'].mean(), inplace=True)

# Calculate correlation matrix after applying rolling mean
correlation_matrix_rolling = qtr_data_rolling[['Home_Index_Rolling', 'SizeRank', 'Year', '30_Yr_Mort_Rate', 'GDP', 'CPI', 'Unemp_Rate', 'Per_Inc', 'Population']].corr()

# Plot the correlation heatmap for rolling mean approach
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix_rolling, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix After Applying Rolling Mean')
plt.show()
```

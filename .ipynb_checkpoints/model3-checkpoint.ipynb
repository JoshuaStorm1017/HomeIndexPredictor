```python
import pandas as pd

# Load dataset
hi_data = pd.read_csv('hi_data.csv')

# Display the first few rows of the dataset
print(hi_data.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Replace infinite values with NaN
hi_data.replace([np.inf, -np.inf], np.nan, inplace=True)

# Selected features for pairplot
selected_features = ['Home_Index', 'GDP', 'CPI', 'Per_Inc', 'Population', 'Unemp_Rate']

# Pairplot for selected features
sns.pairplot(hi_data[selected_features])
plt.show()

# Time series plots for selected features
for column in selected_features[1:]:
    plt.figure()
    hi_data[column].plot(title=column)
    plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Selecting the features for VIF calculation
features = ['GDP', 'CPI', 'Per_Inc', 'Population', 'Unemp_Rate']
X = hi_data[features]

# Calculating VIF
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

# Display VIF values
vif_data
```

```python
# Remove 'CPI' and 'Population' and recalculate VIF
features_reduced = ['GDP', 'Per_Inc', 'Unemp_Rate', 'SizeRank', '30_Yr_Mort_Rate', 'Year_Quarter', 'RegionName_Target_Encoded', 'StateName_Target_Encoded']
X_reduced = hi_data[features_reduced]

# Calculating VIF for the reduced set of features
vif_data_reduced = pd.DataFrame()
vif_data_reduced['Feature'] = X_reduced.columns
vif_data_reduced['VIF'] = [variance_inflation_factor(X_reduced.values, i) for i in range(len(X_reduced.columns))]

# Display VIF values
print(vif_data_reduced)
```

```python
# Remove 'GDP' and recalculate VIF
features_reduced_gdp = ['Per_Inc', 'Unemp_Rate', 'SizeRank', '30_Yr_Mort_Rate', 'Year_Quarter', 'RegionName_Target_Encoded', 'StateName_Target_Encoded']
X_reduced_gdp = hi_data[features_reduced_gdp]

# Calculating VIF for the reduced set of features
vif_data_reduced_gdp = pd.DataFrame()
vif_data_reduced_gdp['Feature'] = X_reduced_gdp.columns
vif_data_reduced_gdp['VIF'] = [variance_inflation_factor(X_reduced_gdp.values, i) for i in range(len(X_reduced_gdp.columns))]

# Display VIF values
print(vif_data_reduced_gdp)
```

```python
# Remove 'Per_Inc' and recalculate VIF
features_reduced_per_inc = ['GDP','Unemp_Rate', 'SizeRank', '30_Yr_Mort_Rate', 'Year_Quarter', 'RegionName_Target_Encoded', 'StateName_Target_Encoded']
X_reduced_per_inc = hi_data[features_reduced_per_inc]

# Calculating VIF for the reduced set of features
vif_data_reduced_per_inc = pd.DataFrame()
vif_data_reduced_per_inc['Feature'] = X_reduced_per_inc.columns
vif_data_reduced_per_inc['VIF'] = [variance_inflation_factor(X_reduced_per_inc.values, i) for i in range(len(X_reduced_per_inc.columns))]

# Display VIF values
print(vif_data_reduced_per_inc)
```

```python
# Selected features with GDP and Year_Quarter, without Per_Inc
features_selected = ['GDP', 'Unemp_Rate', 'SizeRank', '30_Yr_Mort_Rate', 'Year_Quarter', 'RegionName_Target_Encoded', 'StateName_Target_Encoded']
X_selected = hi_data[features_selected]

# Calculating VIF for the selected set of features
vif_data_selected = pd.DataFrame()
vif_data_selected['Feature'] = X_selected.columns
vif_data_selected['VIF'] = [variance_inflation_factor(X_selected.values, i) for i in range(len(X_selected.columns))]

# Display VIF values
print(vif_data_selected)
```

```python
from sklearn.preprocessing import StandardScaler

# Features to be standardized
features_to_standardize = ['GDP', 'Unemp_Rate', 'SizeRank', '30_Yr_Mort_Rate', 'Year_Quarter', 'RegionName_Target_Encoded', 'StateName_Target_Encoded']

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the features
hi_data[features_to_standardize] = scaler.fit_transform(hi_data[features_to_standardize])

# Display the first few rows of the standardized data
print(hi_data.head())
```

```python
# Correct way to select multiple columns
hi_data_new = hi_data[['Home_Index', 'RegionID', 'GDP', 'Unemp_Rate', 'SizeRank', '30_Yr_Mort_Rate', 'Year_Quarter', 'RegionName_Target_Encoded', 'StateName_Target_Encoded']]

# Display the first few rows of the new DataFrame
print(hi_data_new.head())
```

```python
hi_data_new.to_csv('hi_data_new.csv', index=False)
```

```python
hi_data_new.corr()
```

```python
# Standardized split points
train_split_point_std = 0.13
val_split_point_std = 0.89

# Split data using the standardized Year_Quarter values
train_data = hi_data_new[hi_data_new['Year_Quarter'] <= train_split_point_std]
val_data = hi_data_new[(hi_data_new['Year_Quarter'] > train_split_point_std) & (hi_data_new['Year_Quarter'] <= val_split_point_std)]
test_data = hi_data_new[hi_data_new['Year_Quarter'] > val_split_point_std]

# Display the sizes of the splits
print(f'Training data size: {train_data.shape}')
print(f'Validation data size: {val_data.shape}')
print(f'Test data size: {test_data.shape}')
```

```python
train_data.to_csv('train_data_new_std.csv', index=False)
```

```python
val_data.to_csv('val_data_new_std.csv', index=False)
```

```python
test_data.to_csv('test_data_new_std.csv', index=False)
```
